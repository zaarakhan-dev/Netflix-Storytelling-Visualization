# Netflix Content Analysis: Storytelling Through Data Visualization

### Import Libraries

```python
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```



The required libraries are imported for data manipulation and visualization.

- Pandas is used for handling the dataset.
- Matplotlib is used for creating plots.
- Seaborn is used to create attractive statistical visualizations.

### Load the Dataset

```python
df= pd.read_csv("netflix_titles.csv")

df.head()
```

```python
df.info()
```



The Netflix Movies and TV Shows dataset is loaded into a Pandas DataFrame.

The dataset contains information about Netflix content, including title, type, country, release year, rating, genre, and description.

### Data Cleaning

```python
df.isnull().sum()
```

```python
df['director']=df['director'].fillna("Unknown")
df['country']=df['country'].fillna("Unknown")
df['cast']=df['cast'].fillna("Unknown")
```

Missing values were identified and replaced with "Unknown" where appropriate to ensure consistent analysis and visualization.

### Visualization 1

```python
type_counts =df['type'].value_counts()

plt.figure(figsize=(8,5))
plt.bar(type_counts.index, type_counts.values)
plt.title("Movies vs TV shows")
plt.xlabel("Content Type")
plt.ylabel("Count")
plt.show()
```

### Insight

Netflix contains significantly more Movies than TV Shows. This indicates that movies make up the majority of Netflix's content library.

### Visualization 2

```python
country=df['country'].value_counts().head(10)

plt.figure(figsize=(10,5))
plt.bar(country.index,country.values)
plt.title("Top 10 Countries Producing Content on Netflix")
plt.xticks(rotation=45)
plt.xlabel("Country")
plt.ylabel("Count")
plt.show()
```

### Insight

The United States contributes the largest number of titles on Netflix, followed by India and other countries. This highlights Netflix's strong presence in these regions.

### Visualization 3

```python
rating = df['rating'].value_counts()

plt.figure(figsize=(8,5))
plt.bar(rating.index, rating.values)
plt.xticks(rotation=45)
plt.title("Distribution of Content Ratings")
plt.show()
```

### Insight

TV-MA is the most common content rating on Netflix, indicating that a large portion of Netflix's library is intended for mature audiences.

### Visualization 4

```python
year= df['release_year'].value_counts().sort_index()

plt.figure(figsize=(12,5))
plt.plot(year.index,year.values)
plt.title("Netflix Content Release Year")
plt.xlabel("Release Year")
plt.ylabel("Number of Titles")
plt.show()
```

### Insight

The number of titles released increased steadily over time, with significant growth after 2015, demonstrating Netflix's rapid expansion of its content library.

### Visualization 5

```python
genre= df['listed_in'].str.split(", ").explode()
top_genres=genre.value_counts().head(10)

plt.figure(figsize=(10,5))
plt.bar(top_genres.index,top_genres.values)
plt.xticks(rotation=60)
plt.title("Top 10 Genres on Netflix")
plt.show()
```

### Insight

Drama, International Movies, and Comedy are among the most common genres available on Netflix, reflecting the platform's focus on diverse entertainment options.

# Story Behind the Data

The visualizations collectively reveal the evolution of Netflix's content library. Movies dominate the platform, while the United States remains the leading contributor of content. Netflix has experienced rapid growth in its catalog over recent years, with a significant increase in titles after 2015. Most content is targeted toward mature audiences, and drama-related genres are the most prevalent. Overall, the analysis highlights Netflix's expansion into a global entertainment platform offering a diverse range of movies and TV shows.
